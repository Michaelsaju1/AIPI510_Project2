### Step 1: Hypothesis Setting

Research Question: Is the mean of renewable energy sources higher for Balancing Authorities with Microsoft Azure Data Centers.

Null Hypothesis (H₀): There is no difference in the mean of renewable energy sources for Balancing Authorities with known Microsoft Azure Data Centers compared to the Balancing Authorities without known Microsoft Azure Data Centers.

Alternative Hypothesis (H₁): The mean of renewable energy sources for Balancing Authorities with known Microsoft Azure Data Centers is higher compared to the Balancing Authorities without known Microsoft Azure Data Centers.

### Step 2: Choosing Statistical Test

This is a Independent samples t-test (one-sided).
Rationale: Comparing means of a renewable energy sources (MW) between two independent groups (treatment vs. control).
We are specifically interested in an increase amount of usage for BAs with known Microsoft Data Centers, not just a difference.
A one-sided hypothesis is justified if there is no theoretical reason to expect worse means renewable energy sources if there are known Micrsoft Azure Data centers.

### Step 3. Literature Review for Effect Size

Since we are doing simple group and mean comparisons between two groups, we will be using Cohen's d.

For right now we are assuming d = 0.8.

### Step 4.  Power Analysis

```python
from statsmodels.stats.power import TTestIndPower

# Parameters
effect_size = 0.8
alpha = 0.05
power = 0.80

# Power analysis
analysis = TTestIndPower()
sample_size = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, alternative='larger')
print(sample_size)
```

Result:
We need 64 participants per group (total of 128 participants) to detect a medium effect size with 80% power.
